### Fazendo um relatório ja fornecido .

```python
proj = {'Dia':['Segunda','Terça','Quarta','Quinta','Sexta','Sábado','Domingo'],'Horas Trabalhadas':[6,7,8,6,7,5,4],'Bugs Corrigidos':[3,2,1,4,3,2,1],'Tarefas Concluídas':[5,4,6,4,5,3,2]}
mod4 = pd.DataFrame(proj,index = [0,1,2,3,4,5,6])
mod4
```

#Fazendo o total de horas Trabalhadas .

```python
mod4['Horas Trabalhadas'].sum()
```

#Fazendo a media diaria de horas trabalhadas.

```python
mod4['Horas Trabalhadas'].describe()['mean']
```

#Fazendo o total de bugs corrigidos.

```python
mod4['Bugs Corrigidos'].sum()
```

#fazendo a media diaria de bugs corrigidos.

```python
mod4['Bugs Corrigidos'].describe()['mean']
```

#Fazendo o total de tarefas concluidas.

```python
mod4['Tarefas Concluídas'].sum()
```

#Fazendo media diaria de tarefas concluidas.

```python
mod4['Tarefas Concluídas'].describe()['mean']
```

#produtividade diaria (tarefas concluidas por hora)

```python
mod4['Produtividade Diaria'] = mod4['Tarefas Concluídas'] / mod4['Horas Trabalhadas']
mod4
```

```python
mod4['Produtividade Diaria'] = mod4['Produtividade Diaria'].round(2)
mod4
```
